```python
import matplotlib.pyplot as plt

m = 2
b = 5
x = np.arange(0, 10, 0.5)
y_true = m * x + b
noise = np.random.normal(0, 1, len(x))
y_noisy = y_true + noise

plt.plot(x, y_true, label='Ідеальна функція y=2x+5', color='blue')
plt.scatter(x, y_noisy, label='Згенеровані дані', color='red')
plt.title('Лінійна функція з помилками вимірювань')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

MAE = np.mean(np.abs(y_true - y_noisy))
MSE = np.mean((y_true - y_noisy)**2)
print(f'MAE: {MAE:.3f}, MSE: {MSE:.3f}')

df_results = pd.DataFrame({
    'X': x,
    'Y': y_noisy,
    'Y_hat': y_true,
    'MAE': MAE,
    'MSE': MSE
})
df_results.to_csv('Lab2_results.csv', index=False)
```
